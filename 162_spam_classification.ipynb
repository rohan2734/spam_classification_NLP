```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
corpus=[]

for i in range(0,len(dataset)):
    review=re.sub('[^a-zA-z0-9]',' ',dataset['message'][i])# substituting all the values with blank and replacing
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords_list]# we are performing stemming for every word, if word is not in stopwords_list
    review=' '.join(review)
    corpus.append(review)
```

## why do we use max_features ? 

- there may be so many unique words in a dataset 
- ex:. 10k unique words,  we dont need to add all the unique words, we can use the top frequently occuring words 
- there might be words which are hardly used 

## max features is a important hyper parameter 
- max features is needed for building the vocabulary
